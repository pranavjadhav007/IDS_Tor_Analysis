```python
from sklearn.feature_selection import mutual_info_classif
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()

y = df1['Label.1']
y = le.fit_transform(y)
df1['Src IP'] = le.fit_transform(df1['Src IP'])
df1['Dst IP'] = le.fit_transform(df1['Dst IP'])
```

```python
drop = ['Flow ID', 'Label', 'Label.1', 'Timestamp', 'Flow Bytes/s', 'Flow Packets/s']
X = df1.drop(drop, axis=1)
```

```python
importances = mutual_info_classif(X, y)
```

```python
feat_importances = pd.Series(importances, df1.columns[0:79])
```

```python
feat_importances[(feat_importances > 0.42)]
```
